Vamos a añadir 6 columnas más: TEMP_CABELLO, TEMP_OJOS, TEMP_PIEL, VALOR_CABELLO, VALOR_OJOS Y VALOR_PIEL.
De esta forma, luego será más fácil identificar la paleta que corresponda a cada uno de los casos.


```python
#Definimos los casos de temperatura para cada fila del dataframe 'combinaciones'

temperaturas_cab = []
temperaturas_ojos = []
temperaturas_piel = []

for e in combinaciones.CABELLO:
    if e in calidos:
        temperaturas_cab.append('calido')
    else:
        temperaturas_cab.append('frio')

for e in combinaciones.OJOS:
    if e in calidos:
        temperaturas_ojos.append('calido')
    else:
        temperaturas_ojos.append('frio')
        
for e in combinaciones.PIEL:
    if e in calidos:
        temperaturas_piel.append('calido')
    else:
        temperaturas_piel.append('frio')
```

```python
#Definimos los casos de valores para cada fila del dataframe 'combinaciones'

valores_cab = []
valores_ojos = []
valores_piel = []

for e in combinaciones.CABELLO:
    if e in claro:
        valores_cab.append('claro')
    else:
        valores_cab.append('oscuro')
        
for e in combinaciones.OJOS:
    if e in claro:
        valores_ojos.append('claro')
    else:
        valores_ojos.append('oscuro')
        
for e in combinaciones.PIEL:
    if e in claro:
        valores_piel.append('claro')
    else:
        valores_piel.append('oscuro')
```

```python
#Creamos las columnas e igualamos sus valores a las listas que hemos creado anteriormente.
combinaciones['TEMP_OJOS'] = temperaturas_ojos
combinaciones['TEMP_PIEL'] = temperaturas_piel
combinaciones['TEMP_CAB'] = temperaturas_cab
combinaciones['VALOR_CAB'] = valores_cab
combinaciones['VALOR_OJOS'] = valores_ojos
combinaciones['VALOR_PIEL'] = valores_piel
```

Ahora, debemos diferenciar las diferentes paletas de colores que tendremos. Estas son:
    
   1. **PRIMAVERA** - CÁLIDA, CLARA, BRILLANTE
    
   2. **OTOÑO** - FRÍO, CLARO, SUAVE
   
   3. **VERANO** - CÁLIDO, OSCURO, BRILLANTE
    
   4. **INVIERNO** - FRÍO, OSCURO, BRILLANTE

```python
#Determinamos si el total de cada fila es cálido o frío para poder asignar su paleta
temp = []
for i in combinaciones.index:
    if combinaciones['TEMP_CAB'][i] == 'calido' and ((combinaciones['TEMP_OJOS'][i] == 'calido') or \
                                                     (combinaciones['TEMP_PIEL'][i] == 'calido')):\
        temp.append('calido')
    elif str(combinaciones['TEMP_CAB'][i]) == 'frio' and ((combinaciones['TEMP_OJOS'][i]  == 'frio') or\
                                                     (combinaciones['TEMP_PIEL'][i]  == 'frio')):\
       
        temp.append('frio')
    elif combinaciones['TEMP_OJOS'][i] == 'calido' and (combinaciones['TEMP_CAB'][i] == 'calido' or \
                                                       combinaciones['TEMP_PIEL'][i] == 'calido'):\
        
        temp.append('calido')
    elif combinaciones['TEMP_OJOS'][i] == 'frio' and (combinaciones['TEMP_CAB'][i] == 'frio' or\
                                                      combinaciones['TEMP_PIEL'][i] == 'frio'):\
        
        temp.append('frio')
    elif combinaciones['TEMP_PIEL'][i] == 'calido' and (combinaciones['TEMP_CAB'][i] == 'calido' or\
                                                        combinaciones['TEMP_OJOS'][i] == 'calido'):\
        
        temp.append('calido')
    elif combinaciones['TEMP_PIEL'][i] == 'frio' and (combinaciones['TEMP_OJOS'][i] == 'frio' or\
                                                      combinaciones['TEMP_PIEL'][i] == 'frio'):\
        
        temp.append('frio')

    elif combinaciones['TEMP_OJOS'][i] == 'calido' and (combinaciones['TEMP_CAB'][i] == 'frio' and\
                                                        combinaciones['TEMP_PIEL'][i] == 'frio'):\
        
        temp.append('frio')
    else:
        temp.append('no')

combinaciones['TEMP_TOTAL'] = temp
```

```python
#Determinamos si el total de cada fila es claro u oscuro para poder asignar su paleta
val = []
for i in combinaciones.index:
    if combinaciones['VALOR_CAB'][i] == 'claro' and ((combinaciones['VALOR_OJOS'][i] == 'claro') or \
                                                     (combinaciones['VALOR_PIEL'][i] == 'claro')):\
        val.append('claro')
    elif str(combinaciones['VALOR_CAB'][i]) == 'oscuro' and ((combinaciones['VALOR_OJOS'][i]  == 'oscuro') or\
                                                     (combinaciones['VALOR_PIEL'][i]  == 'oscuro')):\
       
        val.append('oscuro')
    elif combinaciones['VALOR_OJOS'][i] == 'claro' and (combinaciones['VALOR_CAB'][i] == 'claro' or \
                                                       combinaciones['VALOR_PIEL'][i] == 'claro'):\
        
        val.append('claro')
    elif combinaciones['VALOR_OJOS'][i] == 'oscuro' and (combinaciones['VALOR_CAB'][i] == 'oscuro' or\
                                                      combinaciones['VALOR_PIEL'][i] == 'oscuro'):\
        
        val.append('oscuro')
    elif combinaciones['VALOR_PIEL'][i] == 'claro' and (combinaciones['VALOR_CAB'][i] == 'claro' or\
                                                        combinaciones['VALOR_OJOS'][i] == 'claro'):\
        
        val.append('claro')
    elif combinaciones['VALOR_PIEL'][i] == 'oscuro' and (combinaciones['VALOR_OJOS'][i] == 'oscuro' or\
                                                      combinaciones['VALOR_PIEL'][i] == 'oscuro'):\
        
        val.append('oscuro')

    elif combinaciones['VALOR_OJOS'][i] == 'claro' and (combinaciones['VALOR_CAB'][i] == 'oscuro' and\
                                                        combinaciones['VALOR_PIEL'][i] == 'oscuro'):\
        
        val.append('oscuro')
    else:
        val.append('no')

combinaciones['VALOR_TOTAL'] = val
```

```python
#Determinamos si el total de cada fila es suave o brillante para poder asignar su paleta
cr = []
for i in combinaciones.index:
    if combinaciones['CROMA_CAB'][i] == 'brillante' and ((combinaciones['CROMA_OJOS'][i] == 'brillante') or \
                                                     (combinaciones['CROMA_PIEL'][i] == 'brillante')):\
        cr.append('brillante')
    elif str(combinaciones['CROMA_CAB'][i]) == 'suave' and ((combinaciones['CROMA_OJOS'][i]  == 'suave') or\
                                                     (combinaciones['CROMA_PIEL'][i]  == 'suave')):\
       
        cr.append('suave')
    elif combinaciones['CROMA_OJOS'][i] == 'brillante' and (combinaciones['CROMA_CAB'][i] == 'brillante' or \
                                                       combinaciones['CROMA_PIEL'][i] == 'brillante'):\
        
        cr.append('brillante')
    elif combinaciones['CROMA_OJOS'][i] == 'suave' and (combinaciones['CROMA_CAB'][i] == 'suave' or\
                                                      combinaciones['CROMA_PIEL'][i] == 'suave'):\
        
        cr.append('suave')
    elif combinaciones['CROMA_PIEL'][i] == 'brillante' and (combinaciones['CROMA_CAB'][i] == 'brillante' or\
                                                        combinaciones['CROMA_OJOS'][i] == 'brillante'):\
        
        cr.append('brillante')
    elif combinaciones['CROMA_PIEL'][i] == 'suave' and (combinaciones['CROMA_OJOS'][i] == 'suave' or\
                                                      combinaciones['VALOR_PIEL'][i] == 'suave'):\
        
        cr.append('suave')

    elif combinaciones['CROMA_OJOS'][i] == 'brillante' and (combinaciones['CROMA_CAB'][i] == 'suave' and\
                                                        combinaciones['CROMA_PIEL'][i] == 'suave'):\
        
        cr.append('suave')
    else:
        cr.append('no')


combinaciones['CROMA_TOTAL'] = cr
```

Ahora, debemos asignar las paletas de colores. Para ello, crearemos la columna 'PALETA'

```python
#Vamos a crear listas que contengan la temperatura, valor y croma de cada paleta.
prim = ['calido', 'claro', 'brillante']
oto = ['frio', 'claro', 'suave']
ver = ['calido', 'oscuro', 'brillante']
inv = ['frio', 'oscuro', 'brillante']
```

```python
#Ahora, en función de los valores por cada columna, debemos asignar la paleta PRIMAVERA, OTOÑO, INVIERNO o VERANO.
paletas = []

for i in combinaciones.index:

    if (combinaciones['TEMP_TOTAL'][i] in prim) and (combinaciones['VALOR_TOTAL'][i] in prim\
                                                     or combinaciones['CROMA_TOTAL'][i] in prim):\
        paletas.append('PRIMAVERA')

    elif (combinaciones['TEMP_TOTAL'][i] in oto) and (combinaciones['VALOR_TOTAL'][i] in oto\
                                                      or combinaciones['CROMA_TOTAL'][i] in oto):\
        paletas.append('OTOÑO')
        
    elif (combinaciones['TEMP_TOTAL'][i] in ver) and (combinaciones['VALOR_TOTAL'][i] in ver\
                                                     or combinaciones['CROMA_TOTAL'][i] in ver):\
        paletas.append('VERANO')

    elif (combinaciones['TEMP_TOTAL'][i] in inv) and (combinaciones['VALOR_TOTAL'][i] in inv\
                                                     or combinaciones['CROMA_TOTAL'][i] in inv):\
        paletas.append('INVIERNO')
        
    elif (combinaciones['VALOR_TOTAL'][i] in oto) and (combinaciones['TEMP_TOTAL'][i] in oto\
                                                     or combinaciones['CROMA_TOTAL'][i] in oto):\
        paletas.append('OTOÑO')

    elif (combinaciones['VALOR_TOTAL'][i] in ver) and (combinaciones['TEMP_TOTAL'][i] in ver\
                                                     or combinaciones['CROMA_TOTAL'][i] in ver):\
        paletas.append('VERANO')
        
    elif (combinaciones['VALOR_TOTAL'][i] in prim) and (combinaciones['TEMP_TOTAL'][i] in prim\
                                                     or combinaciones['CROMA_TOTAL'][i] in prim):\
        paletas.append('PRIMAVERA')
        
    elif (combinaciones['VALOR_TOTAL'][i] in inv) and (combinaciones['TEMP_TOTAL'][i] in inv\
                                                     or combinaciones['CROMA_TOTAL'][i] in inv):\
        paletas.append('INVIERNO')
        
    elif (combinaciones['CROMA_TOTAL'][i] in prim) and (combinaciones['TEMP_TOTAL'][i] in prim\
                                                     or combinaciones['VALOR_TOTAL'][i] in prim):\
        paletas.append('PRIMAVERA')
        
    elif (combinaciones['CROMA_TOTAL'][i] in oto) and (combinaciones['VALOR_TOTAL'][i] in oto\
                                                     or combinaciones['TEMP_TOTAL'][i] in oto):\
        paletas.append('OTOÑO')
        
    elif (combinaciones['CROMA_TOTAL'][i] in ver) and (combinaciones['VALOR_TOTAL'][i] in ver\
                                                     or combinaciones['TEMP_TOTAL'][i] in ver):\
        paletas.append('VERANO')
        
    elif (combinaciones['CROMA_TOTAL'][i] in inv) and (combinaciones['TEMP_TOTAL'][i] in inv\
                                                     or combinaciones['VALOR_TOTAL'][i] in inv):\
        paletas.append('INVIERNO')
    else:
        paletas.append('no')  
```

```python
combinaciones['PALETAS'] = paletas
```
